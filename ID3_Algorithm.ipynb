```python
import numpy as np
import pandas as pd
from math import log
```

```python
df =pd.read_csv("Book1.csv")
df.rename(columns={'Buys computer(class)':'Class','Credit rating':'CreditRating'}, inplace=True)
print(df)
```

```python
CountYes = 0
for i in df['Buys_Computer']:
  if i == 'Yes':
    CountYes += 1
probYes = CountYes/df['Buys_Computer'].count()
probNo = 1-probYes
info = -probYes * log(probYes,2) - probNo * log(probNo,2) 
print(info)
```
